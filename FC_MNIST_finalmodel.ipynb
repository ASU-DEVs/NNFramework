```
import Terminatetensorflow as TTF
```

```
trainFeatures, trainLabels, testFeatures, testLabels, Dataset = TTF.preprocessing_online('MNIST')
trainLabels_ = trainLabels
trainLabels = TTF.labels_to_onehot(trainLabels)
```

```
X = trainFeatures.T
Y = trainLabels
layers_dims = [X.shape[0], 128, 10]
activation_fns = ["sigmoid", "softmax"]
batch_size = 30
epochs = 30
learning_rate = 0.5
costt = "multiclass"
metrics = ["accuracy", "f1 score", "precision","recall","confusion matrix", "macro"]

optimizer = "momentum"

TTF.plotDataset(trainFeatures, Dataset)

model = TTF.Model(layers_dims, activation_fns)

metrics_output = model.fit(X, Y, learning_rate, metrics, epochs, batch_size, costt, optimizer)

TTF.visualizeMetrics(metrics_output, epochs)

TTF.saver.save(model)    
# loaded_model = saver.restore()
  

pred = model.predict(testFeatures.T)
Y_hat = model.predict(trainFeatures.T)
accuracy = TTF.accuracy_metrics(pred.T, testLabels.T) * 100
print(f"\nThe accuracy rate of TEST DATASET is: {accuracy:.2f}%.")
    
```
